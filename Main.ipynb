```python
# Colab 환경에서 GitHub에 있는 CSV 파일 분석 및 합치기

!pip install pandas

# 필요한 라이브러리 설치
import pandas as pd
import os

# GitHub 파일 URL (링크 수정)
csv_files = [
    "https://raw.githubusercontent.com/diligencefrozen/dcinside-data/refs/heads/main/post-dataset/FrozenGallery_PostDataset(14gall_realMain_1).csv",
    "https://raw.githubusercontent.com/diligencefrozen/dcinside-data/refs/heads/main/post-dataset/FrozenGallery_PostDataset(14gall_realMain_2).csv",

    "https://raw.githubusercontent.com/diligencefrozen/dcinside-data/refs/heads/main/post-dataset/FrozenGallery_PostDataset(19gall_realMain_1).csv",
    "https://raw.githubusercontent.com/diligencefrozen/dcinside-data/refs/heads/main/post-dataset/FrozenGallery_PostDataset(19gall_realMain_2).csv",
    "https://raw.githubusercontent.com/diligencefrozen/dcinside-data/refs/heads/main/post-dataset/FrozenGallery_PostDataset(19gall_realMain_3).csv",
    "https://raw.githubusercontent.com/diligencefrozen/dcinside-data/refs/heads/main/post-dataset/FrozenGallery_PostDataset(19gall_realMain_4).csv"
]

# 합칠 데이터프레임 리스트
dfs = []

# 데이터 다운로드 및 로드
for file_url in csv_files:
    local_file = file_url.split('/')[-1]  # 파일명 추출

    # 파일 다운로드
    !wget -q "{file_url}" -O "{local_file}"
    print(f"{local_file} 다운로드 완료")

    # 파일 존재 및 내용 확인
    if os.path.exists(local_file) and os.path.getsize(local_file) > 0:
        try:
            # 데이터 로드
            df = pd.read_csv(local_file)
            dfs.append(df)  # 데이터프레임 리스트에 추가
            print(f"{local_file} 로드 완료 및 데이터 추가.")
        except pd.errors.EmptyDataError:
            print(f"오류: {local_file} 파일이 비어 있습니다.")
        except Exception as e:
            print(f"오류 발생: {e}")
    else:
        print(f"오류: {local_file} 파일이 다운로드되지 않았거나 내용이 비어 있습니다.")

# 데이터프레임 합치기
if dfs:
    main_df = pd.concat(dfs, ignore_index=True)
    main_csv_name = "main.csv"
    main_df.to_csv(main_csv_name, index=False)
    print(f"모든 데이터를 합쳐 {main_csv_name} 파일로 저장했습니다.")
else:
    print("데이터를 합칠 파일이 없습니다.")
```

## 텍스트 마이닝 (Text Mining), 자연어 처리 (NLP),

## 데이터 클렌징(Data Cleansing) / 데이터 전처리(Data Preprocessing)

## 텍스트 정제 (Text Cleaning)

#### 불필요한 데이터(예: HTML 태그, 특수 문자, 공백 등) 제거.
#### 불용어(Stop Words) 제거.
#### 대소문자 변환, 어간 추출(Stemming), 표제어 추출(Lemmatization) 등.

## 토큰화(Tokenization)

#### 텍스트를 단어 또는 문장 단위로 나누는 과정.

## 빈도 분석(Frequency Analysis)

#### 단어 출현 빈도를 분석하고, 이를 기반으로 자주 언급된 단어를 랭킹으로 정리.

## 시각화(Visualization)

#### 워드클라우드 생성 등을 통해 텍스트 데이터를 직관적으로 표현.

```python
from collections import Counter
import re

# main.csv 파일 읽기
try:
    main_df = pd.read_csv("main.csv")
    print("main.csv 파일 로드 성공. 상단 데이터 출력:")
    print(main_df.head())

    # 텍스트 데이터 추출 ('내용' 컬럼)
    if '내용' in main_df.columns:
        text_data = " ".join(main_df['내용'].dropna().astype(str))
        print(f"텍스트 데이터 추출 완료. 총 단어 수: {len(text_data.split())}")
    else:
        print("'내용' 컬럼이 없습니다. 다른 텍스트 열을 사용하세요.")
        text_data = ""

    # 텍스트 전처리
    if text_data:
        # 특수 문자 제거 및 단어 추출
        words = re.findall(r'\b\w+\b', text_data)

        # 불용어 리스트 (원하는 단어 제외 가능)
        stop_words = set(["그리고", "그", "이", "저", "는", "다", "을", "를", "에", "의", "가", "로", "것"])
        filtered_words = [word for word in words if word not in stop_words]

        # 단어 빈도 계산
        word_counts = Counter(filtered_words)

        # 상위 50개 단어 출력
        most_common_words = word_counts.most_common(50)
        print("\n상위 50개 자주 언급된 단어:")
        for word, count in most_common_words:
            print(f"{word}: {count}")
    else:
        print("텍스트 데이터가 없습니다.")
except FileNotFoundError:
    print("main.csv 파일을 찾을 수 없습니다. 파일이 제대로 저장되었는지 확인하세요.")
except Exception as e:
    print(f"오류 발생: {e}")
```
